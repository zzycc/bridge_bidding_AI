```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
import pandas as pd
from multiprocessing import Pool
from sys import getsizeof
from keras.utils import to_categorical
```

```python
######### Use pandas to read "bridge.csv","combats.csv" file #########

bridge_input_num = 5000

bridge_dfN = pd.read_csv('N.csv')
bridge_dfS = pd.read_csv('S.csv')  
bridge_dfE = pd.read_csv('E.csv')  
bridge_dfW = pd.read_csv('W.csv')

#bridge_dfN = pd.DataFrame(df_N)
#bridge_dfS = pd.DataFrame(df_S)
#bridge_dfE = pd.DataFrame(df_E)
#bridge_dfW = pd.DataFrame(df_W)

bridge_dfN_test = bridge_dfN[:bridge_input_num]
bridge_dfS_test = bridge_dfS[:bridge_input_num]
bridge_dfE_test = bridge_dfE[:bridge_input_num]
bridge_dfW_test = bridge_dfW[:bridge_input_num]

bridge_dfN_test = bridge_dfN_test.drop('lin_file',axis = 1)
bridge_dfN_test = bridge_dfN_test.drop('board_num',axis = 1)
bridge_dfS_test = bridge_dfS_test.drop('lin_file',axis = 1)
bridge_dfS_test = bridge_dfS_test.drop('board_num',axis = 1)
bridge_dfE_test = bridge_dfE_test.drop('lin_file',axis = 1)
bridge_dfE_test = bridge_dfE_test.drop('board_num',axis = 1)
bridge_dfW_test = bridge_dfW_test.drop('lin_file',axis = 1)
bridge_dfW_test = bridge_dfW_test.drop('board_num',axis = 1)


bridge_dfN_test['direction'] = 'N'
bridge_dfS_test['direction'] = 'S'
bridge_dfE_test['direction'] = 'E'
bridge_dfW_test['direction'] = 'W'



bridge_df_test = pd.concat([bridge_dfN_test, bridge_dfS_test], ignore_index=True)
bridge_df_test = pd.concat([bridge_df_test, bridge_dfE_test], ignore_index=True)
bridge_df_test = pd.concat([bridge_df_test, bridge_dfW_test], ignore_index=True)

print(len(bridge_dfN_test.iloc[:,:]))
print(len(bridge_dfS_test.iloc[:,:]))
print(len(bridge_dfE_test.iloc[:,:]))
print(len(bridge_dfW_test.iloc[:,:]))

print(len(bridge_df_test.iloc[:,:]))
```

```python
from keras.utils import np_utils
#train_labels[i][1]:
#combats_df.iloc[:,2]

test_output = [[0 for x in range(1)] for y in range(20000)]

for i in range(len(bridge_df_test.iloc[:,:])):
    split_list = (bridge_df_test.iloc[i,4]).split('|')
    order = int(bridge_df_test.iloc[i,5])-1
    
    if(order+1>len(split_list)):
        test_output[i][0] = 1
        continue
    
    split_list[order] = split_list[order].replace("!", "")
    if(split_list[order]=='p' or split_list[order]=='p!' or split_list[order]=='P' or split_list[order]=='P!'):
        test_output[i][0] = 0
        
    elif(split_list[order]=='d' or split_list[order]=='d!' or split_list[order]=='D' or split_list[order]=='D!'):
        test_output[i][0] = 36
    elif(split_list[order]=='r' or split_list[order]=='r!' or split_list[order]=='R' or split_list[order]=='R!'):
        test_output[i][0] = 37
    temp_num = 0    
    if (len(split_list[order]) == 2 and split_list[order][0].isdigit()):
        temp_num = (int(split_list[order][0])-1)*5
        if(split_list[order][1] == 'C' or split_list[order][1] == 'c'):
            temp_num += 1
        elif(split_list[order][1] == 'D' or split_list[order][1] == 'd' ):
            temp_num += 2
        elif(split_list[order][1] == 'H' or split_list[order][1] == 'h'):
            temp_num += 3
        elif(split_list[order][1] == 'S' or split_list[order][1] == 's'):
            temp_num += 4
        elif(split_list[order][1] == 'N' or split_list[order][1] == 'n'): 
            temp_num += 5
        else:
            temp_num = 0
        test_output[i][0] = temp_num
    

#test_output = np_utils.to_categorical(combats_train_label,2)

test_output = np.asarray(test_output)
print(test_output[0:30][:])
print(test_output)
print(test_output.shape)
print(type(test_output))

df = pd.DataFrame(test_output)
print(df)
```
